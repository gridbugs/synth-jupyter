```python
from matplotlib import pyplot as plt
import math
```

```python
k = 3
x = [x / 100 - 1 for x in range(200)]
y = [math.exp(k * (x - 1)) for x in x]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
class Exp01:
    def __init__(self, sharpness):
        self.sharpness = sharpness
        self.b = 1 / (math.exp(sharpness) - 1)
        self.a = -math.log(self.b) / sharpness
        
    def get(self, x):
        return math.exp(self.sharpness * (x - self.a)) - self.b
    
    def get_range(self, xs):
        return [self.get(x) for x in xs]
```
