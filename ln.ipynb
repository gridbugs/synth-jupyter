```python
from matplotlib import pyplot as plt
import math
```

```python
def ln_paramed(k):
    a = 1 / (math.exp(k) - 1)
    b = -math.log(a) / k
    return lambda x: math.log(x + a) / k + b
```

```python
nsteps = 200
x = [x / nsteps for x in range(nsteps)]
```

```python
[x for x in range(1, 200)]
def plot_paramed(k):
    y = [ln_paramed(k)(x) for x in x]
    fig, ax = plt.subplots()
    ax.plot(x, y)
    plt.show()
```

```python
plot_paramed(10)
```
